```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
df = pd.read_csv("dataset.csv")
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nColumn names in dataset:")
print(df.columns)
```

```python
city_col = 'City'
rating_col = 'Aggregate rating'
city_counts = df[city_col].value_counts()
top_city = city_counts.idxmax()
print(f"\nCity with the highest number of restaurants: {top_city} ({city_counts.max()} restaurants)")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=city_counts.head(10).index, y=city_counts.head(10).values, palette="mako")
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_rating = df.groupby(city_col)[rating_col].mean().sort_values(ascending=False)
print("\nAverage Rating per City:")
print(avg_rating.head(10))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=avg_rating.head(10).index, y=avg_rating.head(10).values, palette="viridis")
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
best_city = avg_rating.idxmax()
print(f"\nCity with the highest average rating: {best_city} ({avg_rating.max():.2f})")
```
