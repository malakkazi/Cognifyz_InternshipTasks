```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
df = pd.read_csv("dataset.csv")
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nColumn names in dataset:")
print(df.columns)
```

```python
lat_col = 'Latitude'
lon_col = 'Longitude'
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df[lon_col], y=df[lat_col], alpha=0.4, s=30)
plt.title("Restaurant Locations (Scatter Plot)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
coords = df[[lat_col, lon_col]].dropna()
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(coords)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df[lon_col], y=df[lat_col], hue=df['Cluster'], palette="tab10", s=40, alpha=0.7)
plt.title("Restaurant Clusters Identified by KMeans")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Cluster")
plt.show()
```
