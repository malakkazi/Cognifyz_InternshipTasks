```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
```

```python
df = pd.read_csv("dataset.csv")
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
print("\nColumn names in dataset:")
print(df.columns)
```

```python
cuisine_col = 'Cuisines'
rating_col = 'Aggregate rating'
```

```python
df[cuisine_col] = df[cuisine_col].astype(str).str.strip()
```

```python
cuisine_combinations = df[cuisine_col].value_counts().head(10)
print("\nTop 10 Cuisine Combinations:")
print(cuisine_combinations)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=cuisine_combinations.values, y=cuisine_combinations.index, palette="muted")
plt.title("Top 10 Cuisine Combinations")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combination")
plt.show()
```

```python
avg_rating_cuisine = df.groupby(cuisine_col)[rating_col].mean().sort_values(ascending=False)
print("\nTop 10 Cuisine Combinations by Average Rating:")
print(avg_rating_cuisine.head(10))
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=avg_rating_cuisine.head(10).values, y=avg_rating_cuisine.head(10).index, palette="coolwarm")
plt.title("Top 10 Cuisine Combinations by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine Combination")
plt.show()
```
